```python
#Define a function that converts squarefeet to number
def convert_sqft_num(x):
    tokens=x.split('-')
    if len(tokens) == 2:
        return (float (tokens[0])+float (tokens[1]))/2
    try:
        return float (x)
    except:
        return None
```

```python
df1['total sqft']=df1['total_sqft'].apply(convert_sqft_num)
```

```python
df1.head()
```

```python
df1.describe()
```

```python
df1['price_per_sqft']=df1['price']*100000 /df1 ['total sqft']
df1['price_per_sqft']
```

# Data visualization

```python
#Bar plot for area type vs price
area=df1['area_type'].head(10)
price=df1['price'].head(10)
plt.xlabel("Area type")
plt.ylabel("Price")
plt.bar(area,price)
plt.show()
```

```python
#Pair plot for feature relationships
sns.pairplot(df1, x_vars=['total_sqft', 'bath', 'balcony'], y_vars=['price'], height=5, aspect=0.5)
plt.title("Feature Relationships")
plt.show()
```

```python
#Histogram plot for distribution of prices
sns.histplot(df1['price'], kde=False)
plt.title("Distribution of Prices")
plt.show()
```

```python
#Box plot for price distribution
sns.boxplot(x='price', data=df1)
plt.title("Distribution of Prices")
plt.xlabel("Price")
plt.show()
```
